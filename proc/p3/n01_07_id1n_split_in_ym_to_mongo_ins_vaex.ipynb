```python
def d2j(d, f):
    with open(f, mode="wt", encoding="utf-8") as file:
        json.dump(d, file, ensure_ascii=False, indent=2)
    return 0
        
def f2d(f):
    with open(f) as file:
        d = json.load(file)
    return d
```

```python
def d2jg(d, f):
    with gzip.open(f, mode="wt", encoding="utf-8") as file:
        json.dump(d, file, ensure_ascii=False, indent=2)
    return 0

def fg2d(f):
    with gzip.open(f) as file:
        d = json.load(file)
    return d
```

```python
def d2p(d, f):
    with open(f, mode="wb") as handle:
        pickle.dump(d, handle, protocol=pickle.HIGHEST_PROTOCOL)
    return 0
        
def p2d(f):
    with open(f, "rb") as handle:
        d = pickle.load(handle)
    return d
```

```python
# days_keyset
# days_tableset

def ld2long(ld):
    ld_long = []
    for d in ld:
        d_days = {k:v for k,v in d.items() if k in days_keyset and v == v}
        d_ex = {k:v for k,v in d.items() if k not in days_keyset}
        for day_i, day_i_val in d_days.items():
            if day_i_val == day_i_val:
                d_ex_c = d_ex.copy()
                d_ex_c["day_info"] = day_i
                d_ex_c["day_val"] = day_i_val
                ld_long.append(d_ex_c)
    return ld_long
        
    
```

```python
def ld2dl(ld):
    #list_of_dict: 
    d0k = ld[0].keys()
    dl = dict()
    for k in d0k:
        dl[k] = []
        
    for d in ld:
        for k,v in d.items():
            dl[k].append(v)
    del ld
    return dl
```

```python
import psutil

def slice_0(arr):
    return np.array([arr[i][0] for i in range(len(arr))])

def get_free_mem():
    mem_info = psutil.virtual_memory()
    free_mem = (mem_info.available - mem_info.cached) /mem_info.total
    return free_mem

def wait4ram(d):
    fm = get_free_mem()
    while fm < 0.15:
        print("free_ram:{}".format(fm))
        print("waiting:"+d["f_out"])
        
        time.sleep(600)
        fm = get_free_mem()
        gc.collect()
    print("free_ram:{}".format(fm))
    print("start:"+d["f_out"])
    
```
