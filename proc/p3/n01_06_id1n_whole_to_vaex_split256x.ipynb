```python
def d2j(d, f):
    with open(f, mode="wt", encoding="utf-8") as file:
        json.dump(d, file, ensure_ascii=False, indent=2)
    return 0
        
def f2d(f):
    with open(f) as file:
        d = json.load(file)
    return d
```

```python
def d2jg(d, f):
    with gzip.open(f, mode="wt", encoding="utf-8") as file:
        json.dump(d, file, ensure_ascii=False, indent=2)
    return 0

def fg2d(f):
    with gzip.open(f) as file:
        d = json.load(file)
    return d
```

```python
def d2p(d, f):
    with open(f, mode="wb") as handle:
        pickle.dump(d, handle, protocol=pickle.HIGHEST_PROTOCOL)
    return 0
        
def p2d(f):
    with open(f, "rb") as handle:
        d = pickle.load(handle)
    return d
```

```python
def rec_div_df(df_org, n, d, count=0):
    
    ts = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")    
    div = 2 ** (n)
    
    sch = d["scheme"]
    tb = d["tbl"]
    ym = d["ym"]
    odr = d["odr"]
    
    # temp_f = "../../../unknown/splits/temp_{0}_{1}_{2}_{3}_{4}_{5}_{6}.parquet".format(ym, sch, tb, odr, count, div, ts)
    # temp_f = "../../../storage/temp/splits/temp_{0}_{1}_{2}_{3}_{4}_{5}_{6}.parquet".format(ym, sch, tb, odr, count, div, ts)
    temp_f = "../../../work/procdat/temp/temp_{0}_{1}_{2}_{3}_{4}_{5}_{6}.parquet".format(ym, sch, tb, odr, count, div, ts)
    
    if count == 0:
        del_flg = False
        df = df_org
    else:
        del_flg = True
        df_org.export(temp_f)
        del df_org

        df = vaex.open(temp_f)
        
    df["grp"] = df["cap"] // div
        
    try:
        if n==0 or len(df) <= 1e+5:
            #df.drop(columns=["grp"])
            for c, df_sub in df.groupby("cap"):
                c0 = c[0]
                count += 1
                yield (c0, count, df_sub)
        else:
            #df["grp"] = df["cap"] // div
            for g, df_sub in df.groupby("grp"):
                count += 1
                yield from rec_div_df(df_org=df_sub,n= n-1,d= d, count=count)
    finally:
        del df
        if del_flg:
            print("org del:"+temp_f)
            os.remove(temp_f)
        gc.collect()
        print("fin_rec_iter_count:{}, rec_n:{}".format(count, n))
```

```python
import psutil

# def slice_0(arr):
#     return np.array([arr[i][0] for i in range(len(arr))])
# int(c + c2, 16)//4
# +"{0:03}".format(c)

def slice_01xs(arr):
    arr2 = np.array([arr[i][:2] for i in range(len(arr))])
    return np.vectorize(lambda c:int(c, 16)//div)(arr2)


def slice_01x(arr):
    arr2 = np.array([arr[i][:2] for i in range(len(arr))])
    return np.vectorize(lambda c:"{0:03}".format(int(c, 16)//div))(arr2)

def get_free_mem():
    mem_info = psutil.virtual_memory()
    free_mem = (mem_info.available - mem_info.cached) /mem_info.total
    return free_mem

def wait4ram(d):
    fm = get_free_mem()
    while fm < 0.33:
        print("free_ram:{}".format(fm))
        print("waiting:"+d["f_whole_hdf5"])
        
        time.sleep(1200)
        fm = get_free_mem()
        gc.collect()
    print("free_ram:{}".format(fm))
    print("start:"+d["f_whole_hdf5"])
    
```
